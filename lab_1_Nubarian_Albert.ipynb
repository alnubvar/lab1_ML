## **Задание 4. Другие графики (не менее 10)**

### Что нужно было сделать:
- Построить разнообразные графики, показывающие, какие признаки влияют на Target.
- Использовать scatter, boxplot, гистограммы (столбцовые и круговые).
- Комментировать каждый график и выявленные особенности.

### Примеры графиков и их интерпретация:
1. Подсчитано количество уникальных значений для каждого признака. 
2. Столбчатая диаграмма Target — аналогично, наглядно сравнивает количество объектов по Target.
3. Boxplot Area vs Target — медиана и разброс Area различаются между классами.
4. Boxplot Perimeter vs Target — аналогично, показывает различие в размерах семян.
5. Scatterplot Area vs Perimeter — визуализирует кластеры классов.
6. Scatterplot Compactness vs Solidity — показывает сочетание признаков для разделения Target.
7. Heatmap корреляций — выявляет признаки с высокой взаимной корреляцией и их связь с Target.
8. Pairplot (Area, Perimeter, Solidity) — наглядно показывает распределение объектов по классам.
9. Violinplot Solidity vs Target — медиана и распределение Solidity по классам.
10. KDE Aspect_Ration — плотность распределения для Target=0 и Target=1.

### Выводы:
- Area, Perimeter, Solidity, Convex_Area — ключевые признаки для классификации.
- Графики подтверждают различие классов Target по этим признакам.
- Наблюдаются кластеры, выбросы и дискретная компонента.

### Комменты по графикам:
- Все графики позволяют визуально выявить признаки, влияющие на Target.
- Распределения, кластеры и медианы на графиках дают понимание структуры датасета.
